```python
# Ex 1.3: Generate the crop array into a new variable, use NumPy array slicing

crop_arr = None  # TODO

crop_arr = img_arr.copy()

# Crop array using NumPy slicing
crop_start_x = 137
crop_end_x = 1239
crop_start_y = 108
crop_end_y = 980

crop_arr = crop_arr[crop_start_y:crop_end_y, crop_start_x:crop_end_x, :]
```
